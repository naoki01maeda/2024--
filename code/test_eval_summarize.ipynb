```python
import numpy as np

def cal_std(one,two,three,four,five):

    all_std = np.zeros((one.shape[0], one.shape[1]))
    all_f1 = np.concatenate([one[np.newaxis,:,:],two[np.newaxis,:,:],three[np.newaxis,:,:],four[np.newaxis,:,:],five[np.newaxis,:,:]])

    for model_num in range(len(all_f1[0])):

        for label_num in range(len(all_f1[0][0])):
            std_list = []
            for split_num in range(len(all_f1)):

                std_list.append(all_f1[split_num][model_num][label_num])
            
            std = np.std(std_list)
            all_std[model_num][label_num] = round(std, 3)
    
    return all_std
```
