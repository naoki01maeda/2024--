```python
import pickle
import os

path = "C:/Users/naoki/anaconda_work/risk_facter_estimation_network_main/dataset/pkl_data/"
files = os.listdir(path)
img_list = []
image_id = 1
limit_id = 4000
for i in range(image_id, len(files)):
  print(image_id + i - image_id)
  if i == limit_id:
    break
  with open(path + '{0:06}.pkl'.format(image_id + i - image_id), 'rb') as web:
    img_list.append(pickle.load(web))#pklファイルのファイル名を img_list に格納
```

# optical_flow(jpg)に変換

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

save_path = "C:/Users/naoki/anaconda_work/risk_facter_estimation_network_main/dataset/optical_flow/"

if os.path.isdir(save_path) == False:
    os.mkdir(save_path)

for img_index, img_data in enumerate(img_list):
    img = img_data['flow_img'].transpose(1, 2, 0)
    img = np.array(img)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    cv2.imwrite(save_path + '{}.jpg'.format(image_id + img_index+1),img)#optical_flow画像
```
